```python
import torch
import numpy as np 
import matplotlib.pyplot as plt

allTimeSeries=torch.load("data/192T/time_series.pt")
allMasks=torch.load("data/192T/masks.pt")
```

```python
timeStepVector = [i for i in range(192)]

timeSeriesOnePatient = allTimeSeries[1]
masksOnePatient = allMasks[1]

for i in range(59):               #59 is num of variables
    oneTimeSeries = np.asarray(timeSeriesOnePatient[..., i])
    oneMask = np.asarray(masksOnePatient[..., i])
    plt.plot(timeStepVector, oneTimeSeries)
    xVector=[]
    yVector=[]
    for j in range(192):
        if (oneMask[j]==0.0):                   #if datapoint originally NOT missing - if data is actually true
            xVector.append(j)
            yVector.append(oneTimeSeries[j])
    print('variable number: ', i)
    plt.scatter(xVector, yVector, marker='o', color="red")
    plt.xlabel('time step')
    plt.ylabel('medical variable')
    plt.show()
```
